# Loops

In python, there are times that we need to tell the programme to repeat itself, or to repeat a specific part of the code. They are normally seen as a block of code. Similar to if/else statements. 

There are many different types of loops, but we will focus on 2 types:
* Counting loops
* Conditional loops

There are also two main loops that will be used: 
* **For loop**
* **While loop**

A **counting loop** is a loop in which you specify how long you want the block of code to repeat itself

Ex. 

```python

  print('hello')
```

## Breaking loops

If you want the loop to stop at a specific point, you can **break** the loop.

This done by using the `break` function.

```python
for x in range(5):
  if x == 2:
    break
  else:
    print(x)
```

This code block means that from 0, 1, 2, 3, 4 if the number is equal to 2, the code will break (stop). But if the number is not equal to 2, it will print the number.

It is important to note that once the condition for the break has been met (finding the number equal to 2) the code will not continue to work after that point.

````python
for x in range(5):
  if x == 2:
    break
  else:
    print(x)
````

## Len()

The function `len()` is used to find the individual number of whatever item you afre using it on. 

For instance, if you are using it on words (strings), you would be finding the number of letters each word has. 

If you are using it on a list, you are finding out the number of items within the list. 


````python
words = ['ball', 'mat', 'rat', 'laptop', 'lead', 'leash', 'mouse', 'screen','charger', 'pod', 'airpod', 'table', 'chair']

for word in words:
  if len(word) > 6:
    break
  else:
    print(word)
````

# While Loops

While loops are used for iterating as long as a condition is True. It is useful for when you need to repeat a piece of code without the number of repetitions needing to be specified. 

Similar to the **for loop** and **if/else statements**, the while loop requires a colon `(:)` and an indented block after to show the block of code that is to be repeated. 

````python
# An example of a while loop in action

counter = 0

while counter < 3:     # While the counter is less than 3
  counter += 1          # Add 1 to the counter 
  print(counter)        # Print the output of the counter then start again until the counter is larger than 3
````

````python
# Write a while loop that adds up all the numbers to the total 100

total = 0

while total < 10:
  total += 1
  print(total)

````

````python
n = 1 

while n < 5:
  print('Hello Python')
  n += 1
````

# Using Elif statements

Elif statements are part of the if-else statements we are already familiar with. However, instead of just having 1 alternative option, `elif` allows us to have multiple other options. 

````python
x = 0

while x < 5:
  if x == 0:
    print(f'x is {x}')
    x += 1
  elif x == 1:
    print(f'x is {x}')
    x += 1
  elif x == 2:
    print(f'x is {x}')
    x += 1
  elif x == 3:
    print(f'x is {x}')
    x += 1
  else:
    break
````

````python
# Adding elements to a list using a 'while loop'

my_list = []
i = 0

while len(my_list) < 4:     # While the length of the list is less than 4
  my_list.append(i)         # Add 'i' to the list
  i += 1                    # Add 1 to i each time you add it to the list

print(my_list)
````

````python
empty_list = []
i = 1


while i < 20:
  empty_list.append(i)
  i += 2

print(empty_list)
````

````python
# Find the sum of numbers in a  list using 'while loop'

my_list = [2, 5, 3, 12, 8]
i = 0
sum = 0

while i < len(my_list):       # While i is less than length of the list
  sum += my_list[i]
  i += 1                      # Increase i by 1 so that it goes through to list
  # print(f'The list is at position {i} and it is adding the number to the sum')           # The sum is the addition of the number in the list

print(sum)
````

````python
# Printing the square of numbers using a while loop

n = 1

while n <= 10:
  square_number = n ** 2
  print(f'{n}-squared: {square_number}')
  n += 1
````

# Functions

A function is a block of code that performs a specific task that can be reused many times. You can pass information and it can send information back.

There are 2 types of fucntions:
* **Built in functions**: These are built into the python programming language.

* **User-defined functions**: These are functions made by users to do a specific task that is relevant to their programmes. 

Basic function syntax 

```python

  # Function Body
  ```
Points to note, when writing functions, they all start with `def` then the functions name and whatever parameters are needed. Similar to loops and if/else statements, there are colons after the first line and the next line is indented. 


  Example 

  ```python

print(f'Hello {name}! Good morning!)
```

Inside the brackets, you pass any parameters. As you can see, this greet function requires a name to be passed. So for the example, the calling of the function will be `greet('Adaobi')`. 

As you can see in the rest of the clode for the function, whatever the name is passed as an argument is placed in the print statement.

## Function inputs

The input of function can be anything that is necessary for the function itself. A number, a string, a bunch of numbers, and other stuff. 

The easiest way to implement the ability to have multiple inputs is using `*args`.

First off to note: 'args' is just a name. In the function, as long as you add the asterisk ( * ), you can name it anything that is related to the function/program.

````python
# This is one way to define the addition of numbers 
def plus(a, b):
  return a + b

plus(4,3)
````

````python
# This is using the *args argument
def plus(*numbers):
  result = 0
  for x in numbers:
    result += x
  return result

plus(12,3,4)
````
